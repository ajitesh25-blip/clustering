1.What is unsupervised learning in the context of machine learning?

Ans. Unsupervised learning is a machine learning method where algorithms analyze and cluster raw, unlabeled datasets. Instead of using predefined answers, the model independently uncovers hidden structures, natural groupings, and relationships within the data without any human guidance.

2. How does K-Means clustering algorithm work?

Ans. K-Means clustering is an unsupervised machine learning algorithm that groups similar data points into a predefined number of non-overlapping clusters (\(K\)) based on their distance from central points called centroids.

3. Explain the concept of a dendrogram in hierarchical clustering.

Ans. A dendrogram is a tree-like diagram that visually maps the step-by-step arrangement and relationships of clusters generated by a hierarchical clustering algorithm. It tracks how individual data points or groups of points are merged (or divided) at various levels of similarity or distance.

4. What is the main difference between K-Means and Hierarchical Clustering?

Ans.The main difference between K-Means and Hierarchical clustering is how the clusters are formed and structured. K-Means divides data into a predetermined number of flat, non-overlapping groups, while Hierarchical clustering builds a tree-like hierarchy of nested clusters (dendrogram) that can be explored at multiple levels of granularity.

5. What are the advantages of DBSCAN over K-Means?

Ans.DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a highly flexible clustering algorithm. Unlike K-Means, it groups data points based on local data density rather than distances from a central point. You can use machine learning tools like Scikit-Learn to implement both algorithms and compare their outcomes.

6. When would you use Silhouette Score in clustering ?

Ans. Finding the Optimal Number of Clusters: When you need to determine the best value for \(k\) (the number of clusters) before running an algorithm, testing different values and plotting their Silhouette Scores will help you find the highest peak, which indicates the best grouping.Validating Cluster Quality: When you want a single, easily interpretable score to objectively grade how distinct and well-separated your groups are.Analyzing High-Dimensional Data: When your dataset has too many dimensions to visualize on a 2D or 3D plot, the Silhouette Score acts as a mathematical proxy to ensure your clusters remain dense and distinct.

7. What are the limitations of Hierarchical Clustering?

Ans. The algorithm requires calculating and storing similarity matrices, resulting in a high time complexity and demanding heavy memory usage. This makes it unscalable and practically unusable for massive, real-time datasets.It is a "greedy" algorithm, meaning it makes locally optimal decisions at every step. Once the algorithm combines two clusters or splits a group, this decision cannot be undone at a later stage, potentially leading to suboptimal final groupings.Outliers and noise points can severely distort distance calculations and warp the structure of the resulting tree-diagram (dendrogram). This can result in misleading interpretations of the data.

8. Why is feature scaling important in clustering algorithms like K-Means?

Ans. Feature scaling transforms dataset variables to a common numeric range. In K-Means, which groups data based on proximity measures like Euclidean distance, scaling ensures that features with larger magnitudes do not disproportionately dominate the cluster assignments. It guarantees every feature contributes equally to similarity calculations.

9. How does DBSCAN identify noise points?

Ans. In the DBSCAN (Density-Based Spatial Clustering of Applications with Noise) algorithm, noise points are identified based on their density relative to two user-defined parameters: Epsilon (\(\epsilon \)), the search radius, and MinPts, the minimum number of points required to form a dense region.

10. Define inertia in the context of K-Means.

Ans. In K-Means clustering, inertia measures the internal cohesiveness of the clusters by calculating the sum of squared distances from each data point to its assigned centroid. It serves as an objective function that the algorithm seeks to minimize; a lower inertia value indicates tighter, more densely packed clusters.

11. What is the elbow method in K-Means clustering?

Ans. The Elbow Method is a heuristic used in K-Means clustering to determine the optimal number of clusters (\(K\)) by plotting the clustering variance against a range of values for \(K\) and locating the "elbow" (a distinct bend or point of inflection) on the resulting curve.

12. Describe the concept of "density" in DBSCAN.

Ans. In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), density refers to how closely packed data points are within a specific region of the feature space. Clusters are defined as continuous regions of high data point density, separated by areas of lower density.

13. Can hierarchical clustering be used on categorical data?

Ans. Yes, hierarchical clustering can be used on categorical data. Because traditional distance measures (like Euclidean distance) do not work for non-numerical information, you must use specialized similarity measures or pre-process the data.

14. What does a negative Silhouette Score indicate?

Ans. A negative Silhouette Score indicates that a data point has been incorrectly assigned to a cluster, meaning it is, on average, closer to the data points in a neighboring cluster than to the points in its own assigned cluster.

15. Explain the term "linkage criteria" in hierarchical clustering.

Ans. In hierarchical clustering, linkage criteria are mathematical rules that determine the distance or dissimilarity between groups (clusters) of data points. They allow algorithms to measure how far apart or similar different sets of observations are, guiding which clusters to merge or split at each step of the analysis.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Ans. K-Means clustering is a centroid-based algorithm that divides a dataset into \(k\) distinct groups by minimizing the squared distance of data points to their assigned cluster centers. It relies on Euclidean distance and inherently assumes that clusters are spherical, uniformly dense, and roughly equal in size.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

Ans. The maximum radius of the neighborhood around any given data point. Two points are considered neighbors if the distance between them is less than or equal to \(\epsilon \).If \(\epsilon \) is too small: Most data points will fail to meet the density criteria and will be mistakenly classified as noise.If \(\epsilon \) is too large: Distinct clusters will merge into a single large cluster, and the algorithm will fail to distinguish between separate dense regions.The threshold number of data points (including the point itself) required to fall within the \(\epsilon \)-radius for a region to be considered dense.If MinPts is too small: The algorithm becomes highly sensitive to minor variations, leading to fragmented clusters and classifying many random points as valid clusters.If MinPts is too large: The algorithm requires highly concentrated groupings to form clusters. Valid clusters with lower densities will be dismissed as noise, meaning more data will be discarded as outliers.

18. How does K-Means++ improve upon standard K-Means initialization?

Ans. The process of selecting initial cluster centers entirely at random from the data points.A probabilistic process where the first center is chosen randomly, but subsequent centers are chosen with a probability proportional to their squared distance from the closest existing center. This mathematically guarantees that initial centers are spread out across the dataset.

19. What is agglomerative clustering?

Ans. Agglomerative clustering is a "bottom-up" hierarchical machine learning algorithm. It begins by treating every single data point as an independent cluster, then iteratively merges the closest pairs of clusters based on distance until all data forms a single, comprehensive group.

20.  What makes Silhouette Score a better metric than just inertia for model evaluation?

Ans. The Silhouette Score is a better metric than inertia for model evaluation because it simultaneously measures both cohesion (how close points are to their own cluster) and separation (how distant points are from neighboring clusters).

```python
'''21.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.'''

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering (4 Centers)")
plt.legend()
plt.show()
```

```python
'''23.Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.'''

import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

dbscan = DBSCAN(eps=0.15, min_samples=5)
labels = dbscan.fit_predict(X)

outlier_mask = labels == -1
clustered_mask = ~outlier_mask

plt.figure(figsize=(8, 6))

plt.scatter(
    X[clustered_mask, 0],
    X[clustered_mask, 1],
    c=labels[clustered_mask],
    cmap="viridis",
    s=50,
    edgecolors="k",
    label="Clusters",
)

plt.scatter(
    X[outlier_mask, 0],
    X[outlier_mask, 1],
    c="red",
    marker="x",
    s=80,
    linewidths=2,
    label="Outliers (Noise)",
)

plt.title("DBSCAN Clustering on Moon Data with Highlighted Outliers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(loc="upper right")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
'''24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each
cluster.'''

import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

wine_data = load_wine()
X = wine_data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(X_scaled)

cluster_sizes = pd.Series(clusters).value_counts().sort_index()
print("Size of each cluster:")
for cluster_id, size in cluster_sizes.items():
    print(f"Cluster {cluster_id}: {size} samples")
```

```python
'''25.Use make_circles to generate synthetic data and cluster it using DBSCAN. Plot the result.'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN

X, _ = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

dbscan = DBSCAN(eps=0.15, min_samples=5)
labels = dbscan.fit_predict(X)

unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

plt.figure(figsize=(8, 6))

for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)
    xy = X[class_member_mask]

    plt.scatter(
        xy[:, 0], xy[:, 1],
        color=tuple(col),
        edgecolor='k',
        s=50,
        label=f'Cluster {k}' if k != -1 else 'Noise'
    )

plt.title('DBSCAN Clustering on Concentric Circles')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
'''31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries.'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, _ = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], s=30, edgecolor='k', alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, marker='X', c='red', label='Centroids')
plt.title("K-Means Decision Boundaries (Overlapping Clusters)")
plt.legend()
plt.show()
```

```python
'''32.Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results.'''

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

def run_pipeline():

    print("Loading Digits dataset...")
    digits = load_digits()
    X = digits.data
    y_true = digits.target

    print("Scaling features...")
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    print("Reducing dimensions with t-SNE...")
    tsne = TSNE(n_components=2, perplexity=30, random_state=42, n_iter=1000)
    X_2d = tsne.fit_transform(X_scaled)

    print("Clustering with DBSCAN...")
    dbscan = DBSCAN(eps=3.5, min_samples=15)
    cluster_labels = dbscan.fit_predict(X_2d)

    n_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
    n_noise = list(cluster_labels).count(-1)
    print(f"Estimated number of clusters: {n_clusters}")
    print(f"Estimated number of noise points: {n_noise}")

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    scatter1 = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=cluster_labels, cmap='tab20', s=15, alpha=0.8)
    plt.colorbar(scatter1, label='DBSCAN Cluster ID')
    plt.title(f'DBSCAN Results ({n_clusters} Clusters Found)\nNoise points marked in dark/grey')
    plt.xlabel('t-SNE 1')
    plt.ylabel('t-SNE 2')

    plt.subplot(1, 2, 2)
    scatter2 = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y_true, cmap='tab10', s=15, alpha=0.8)
    plt.colorbar(scatter2, label='Actual Digit (0-9)')
    plt.title('Ground Truth Digit Labels')
    plt.xlabel('t-SNE 1')
    plt.ylabel('t-SNE 2')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    run_pipeline()
```

```python
'''34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot.'''

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

cancer_data = load_breast_cancer()
X = cancer_data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

k_values = range(2, 7)
inertias = []

for k in k_values:

    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertias, marker='o', linestyle='-', color='b', linewidth=2)
plt.title('K-Means Inertia (Breast Cancer Dataset)', fontsize=14)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
plt.xticks(k_values)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
'''43.Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
side-by-side.'''

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

X, _ = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

X = StandardScaler().fit_transform(X)

kmeans = KMeans(n_clusters=2, n_init='auto', random_state=42)
km_labels = kmeans.fit_predict(X)

dbscan = DBSCAN(eps=0.3, min_samples=5)
db_labels = dbscan.fit_predict(X)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.scatter(X[:, 0], X[:, 1], c=km_labels, cmap='viridis', edgecolors='k', s=30)
ax1.set_title("KMeans Clustering\n(Fails on Nested Shapes)")

ax2.scatter(X[:, 0], X[:, 1], c=db_labels, cmap='plasma', edgecolors='k', s=30)
ax2.set_title("DBSCAN Clustering\n(Successfully Finds Rings)")

plt.tight_layout()
plt.show()
```

```python
'''44.Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.'''

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import silhouette_samples, silhouette_score

iris = load_iris()
X = iris.data

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X)

silhouette_vals = silhouette_samples(X, cluster_labels)
avg_score = silhouette_score(X, cluster_labels)

fig, ax = plt.subplots(figsize=(8, 6))
y_lower = 10

for i in range(n_clusters):
    ith_cluster_vals = silhouette_vals[cluster_labels == i]
    ith_cluster_vals.sort()

    size_cluster_i = ith_cluster_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    color = cm.nipy_spectral(float(i) / n_clusters)
    ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_vals,
                      facecolor=color, edgecolor=color, alpha=0.7)

    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

ax.axvline(x=avg_score, color="red", linestyle="--", label=f'Avg: {avg_score:.2f}')
ax.set_title("Silhouette Plot for Iris Dataset (k=3)")
ax.set_xlabel("Silhouette Coefficient Values")
ax.set_ylabel("Cluster Label")
plt.legend()
plt.show()
```

```python
'''46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features).'''

import pandas as pd
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

wine_data = load_wine()
X = wine_data.data
feature_names = wine_data.feature_names

df = pd.DataFrame(X, columns=feature_names)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
df['Cluster'] = kmeans.fit_predict(X_scaled)

subset_df = df.iloc[:, :4].copy()
subset_df['Cluster'] = df['Cluster']

sns.pairplot(
    subset_df,
    hue='Cluster',
    palette='viridis',
    diag_kind='kde',
    plot_kws={'alpha': 0.7}
)

plt.suptitle("K-Means Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
'''47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count.'''

import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

X, _ = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

rng = np.random.RandomState(42)
noise = rng.uniform(low=-15, high=15, size=(50, 2))
X = np.vstack([X, noise])

db = DBSCAN(eps=1.2, min_samples=5).fit(X)
labels = db.labels_

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points found: {n_noise}")
```
